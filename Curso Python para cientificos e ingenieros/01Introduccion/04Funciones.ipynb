# **Ejercicio**

Escribir una funcion que recibe como argumento un numero
En un rango entre 0 y ese numero (sin contar el ultimo)
Si el numero es multiplo de 3, entonces Mostrar un mensaje "Es multiplo de 3"
Si no es asi, pero es multiplo de 5, entonces mostrar un mensaje analogo al anterior.
Por ultimo si no es multiplo de ninguno, mostrar en mensaje el numero.

```
def mi_fun(n=5):
  """Determina si un numero n  es multiplo de 3 o 5"""
  for numero in range(n):
    if numero % 3 is 0:
      print("{nu} Es multiplo de 3".format(nu = numero))
    elif numero % 5 == 0:
      print("{0} Es multiplo de 5".format(numero))
    else:
      print(numero)

mi_fun(8)
```
