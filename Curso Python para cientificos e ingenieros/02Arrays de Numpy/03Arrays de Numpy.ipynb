# **Operaciones con arrays**

Las funciones universales (nfunc) operan sobre arrays de NumPy elemento a elemento y siguiendo las reglas de broadcasting.

*  Funciónes matemáticas: `sin, cos, sqrt, exp`, ..
*  Operaciones lógicas <, ~, ..
*  Funciones lógicas: `all, any, isnan, allclose`,

*Nota: Las funciones matemáticas siempre devuelven el mismo tipo de datos de entrada*.

```
import numpy as np
```

```
a = np.arange(2 * 3).reshape(2, 3)
a
```

```
np.sqrt(a)
```

```
np.sqrt(np.arange(-3,3))
```

```
a = np.arange(-3,3).astype(complex)
a
```

```
np.sqrt(a)
```

**Funciones lógicas**

Las comparaciones devuelven un array de booleanos.

```
a = np.arange(6)
b = np.ones(6).astype(int)
a, b
```

```
#Comparar dos arrays a simple vista no tiene sentido, sin embargo la operacion de
#comparación esta vectorizada, es decir se realiza elemento a elemento.
a < b
```

```
np.any(a < b)  #Devuelve True si algun valor del array es verdadero
```

```
np.all(a < b) #Devuelve False, de forma analoga a la funcion any
```

```
a = np.arange(6).astype(float)
b = np.ones(6)
a, b
```

```
np.isclose(a, b, rtol=1e-6)
```

```
np.allclose(a, b, rtol=1e-6)
```

**Ejercicio 3.**

*  Crea una matriz aleatoria 5x5 y halla los valores minimo y máximo.
*  Normaliza esa matriz entre 0 y 1

```
a = np.random.randint(10, 99, (5, 5))
min = np.amin(a)
max = np.amax(a)
print(a)
print("\nMinimo es: {0}".format(min))
print("Minimo es: {0}".format(max))
a -= min
b = a / max
print(b)
b.size
```

```
import matplotlib.pyplot as plt
#x = np.reshape(np.around(b,decimals=2),(1, 25))
matriz_redondeada = np.round_(b,decimals=1)
print(matriz_redondeada)
X = np.histogram(matriz_redondeada, bins=np.arange(10),density=True, range=(min,max))

plt.hist(X)
#plt.ylim(0,1)
plt.grid()
plt.show()
```
