## **¿Qué es matplotlib?**



*   Estándar de *facto* para visualización en Python
*   Pretende ser similar a las funciones de visualización de MATLAB.
*   Diferentes formas de usarla: interfaz `pyplot` y/o utilizando el paradigma POO.

Lo primero que tenemos que hacer es activar el modo inline. de esta manera las figuras aparecerán automaticamente incrustadas en el notebook.



```
%matplotlib inline
```

Importamos los paquetes necesarios.

```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
print('matplotlib: {}'.format(matplotlib.__version__))
```

## **Interfaz pyplot**


*   Función plot: listas o arrays.
*   Personalización de líneas, colores, leyendas
*   Otros tipos de gráficas: `scatter, contour, fill_between`
*   Comó pintar una función en 1D, en 2D, 3D.
*   `plt.savefig`



**Trabajando con las listas de python, no es muy efeciente**

```
plt.plot([1, -1, 0, 3, -1]) #Grafica los valores de la lista CONTRA las posiciones de la lista
plt.grid()
#posiciones =  eje "x"
#valores de lista = eje "y"
```

```
plt.plot([0.0, 0.1, 0.2, 0.5, 0.6], [1, -1, 0, 3, -1]) #De forma explicita tenemos dos listas de python
#Donde la primera hace referencia a el eje X y la segunda lista al eje Y
```

**Utilizando los arrays de NumPy**, es donde matplotlib, comienza a ser una potente herramienta.
Ejemplo, si nosotros queremos graficar la siguiente función.

```
%%latex
$$f(x) = e^{-x^2}$$
```

```
def f(x):
    return np.exp(-x ** 2)
```

```
x = np.linspace(-1, 5, num=30)
```

```
#Consultar mas en: https://matplotlib.org/3.2.2/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot
plt.plot(x, f(x),'sm',label="Función f(x)")
plt.xlabel('Eje x')
plt.ylabel('$f(x)$')
plt.legend() #Mostramos la "label" definida en la funcion plot
plt.grid()  #Nos muestra la regilla o cuadricula, tipo EXCEL
plt.title('$f(x) = e^{-x^2} $')
plt.show()
```
