# **Interfaz orientada a objetos**

```
%matplotlib inline
```

```
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(x):
    return np.exp(-x ** 2)
```

```
x = np.linspace(-1, 5, num=30)
x
```

```
fig, axes = plt.subplots()

axes.plot(x,f(x), 'ro', label="Funcion f(x)")  #Ahora graficas con el objeto axes
axes.set_xlim(-2, 4)  #por defecto se ajusta al dominio donde se evalua la funcion
axes.set_ylim(-1, 2) #por defecto se ajusta al valor minimo y maximo de la funcion
axes.grid(True)
fig.savefig("Grafica1.png")  #guardamos la grafica, con ayuda del objeto fig
```

Que pasa si queremos obtener una figura, con varios sistemas de ejes.

```
fig, axes = plt.subplots(1, 2) #en una fila quiero 2 axes

axes[0].plot(x, f(x))
axes[0].set_xlabel('Eje x')

axes[1].plot(x, -f(x), 'sm')
axes[1].set_xlabel('Eje x de grafica derecha')
plt.show()
```

```
axes  #Nos muestra nos objetos numpy y cada uno con un objeto axes diferente
```

```
fig, axes = plt.subplots(1, 2, sharey=True) #en una fila quiero 2 axes
 #el argumento sharey hace que compartan eje "y" las graficas
axes[0].grid(True)
axes[0].plot(x, f(x), color='green')
axes[0].set_xlabel('Eje x')

axes[1].plot(x, -f(x), 'sm')
axes[1].set_xlabel('Eje x de grafica derecha')
```


Con s y c podemos modificar el tamaño y el color respectivamente. Para el color, a cada valor numérico se le asigna un color a través de un mapa de colores; ese mapa se puede cambiar con el argumento cmap. Esa correspondencia se puede visualizar llamando a la función colorbar.

```
x = np.random.randn(100)
y = np.random.randn(100)

s = 200 * np.random.randn(100)
c = np.random.randn(100)  #array para el mapa de colores
plt.scatter(x, y, s, c, cmap=plt.cm.gist_rainbow)
plt.colorbar()
plt.grid()
```
