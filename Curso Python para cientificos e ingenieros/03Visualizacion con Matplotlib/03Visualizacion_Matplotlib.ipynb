```
import numpy as np
import matplotlib.pyplot as plt
```

```
%%latex
$$g(x, y) = \cos(x) + \sin(y)^2 $$
```

```
def g(x, y):
    return np.cos(x) + np.sin(y) ** 2
```

```
x = np.linspace(-2, 2)
y = np.linspace(-2, 2)

X, Y = np.meshgrid(x, y)
```

```
Z = g(X, Y)
```

```
#Alternativa, pero no la recomendada.
plt.contour(X, Y, Z)
plt.colorbar()
```

```
fig, axes = plt.subplots()

axes.contour(X, Y, Z,cmap=plt.cm.autumn)
axes.grid(True)
```

```
fig, axes = plt.subplots()
axes.contourf(X, Y, Z,cmap=plt.cm.autumn)
axes.grid(False)
```

```
fig, axes = plt.subplots()
axes.contourf(X, Y, Z, np.linspace(-3.2, 1),cmap=plt.cm.autumn)
axes.grid(False)
```

```
datos = np.loadtxt('temperaturas.csv', usecols=(1, 2, 3), skiprows=1, delimiter=',')
```

```
fig, axes = plt.subplots()

x = np.arange(len(datos[:, 1]))

temp_media = (datos[:, 1] + datos[:, 2]) / 2

axes.plot(x, datos[:, 1], 'r') #temperaturas maximas
axes.plot(x, datos[:, 2], color='b') #temperaturas minimas
axes.plot(x, temp_media, color='k') #temperaturas media
axes.grid()
```
