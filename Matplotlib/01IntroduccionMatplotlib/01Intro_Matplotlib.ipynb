# **Matplotlib**

Es una **libreria de visualización** más popular de Python (desde el año *2003*, creada por: **Jonh D. Hunter**).

*  En un principio se creo para simular el comportamiento gráfico de **MATLAB**.
*  Hoy en día matplotlib es la referencia para otras librerias de visualización como: *seaborn, ggplot, Bokeh, Plotly*. etc.
*  **`Pandas`** posee el método plot (importado de matplotlib), que permite crear y personalizar con cierta facilidad una gráfica con los datos contenidos en el DataFrame.

Es una libreria muy potente y extensa, que permite la creación de visualizaciones 2D, 3D, y mapas.

*  Matplotlib depende a su vez de NumPy (Especializada en el calculo de arrays mutidimencionales).
*  La mayoria de las funciones o metodos de *matplotlib* estan en la sublibreria **`pyplot`**, que suele importarse con el alias de **`plt`**:

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## **Objetos basicos de matplotlib**

Los objetos basicos, que se deben conocer de matplotlib

**1.figure** (figura): Es un contendor, como el nombre lo indica, sirve para guardar todas las graficas que generen dentro de ella, "*es como el marco de una fotografica*".

ref: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html

```python
#Tiene varios argumentos, destacamos el color de fondo facecolor, num es el numero de figura
plt.figure(num=1,facecolor='paleturquoise',frameon=True)
#El método title, nos permite rotular nuestra figura.
plt.title("Figura")
#El método show es necesario para mostrar nuestra figura
plt.show()
```

**2.axes** (conjunto de ejes): Las líneas ortogonales que definen las referencias para las dos dimensiones involucradas, matematicamente hablando el "eje cartesiano"

Hay que mencionar que es posible tener varios AXES (plural) en una Figura.
axis: es un eje (singular)

<img src="https://www.interactivechaos.com/sites/default/files/inline-images/tutorial_matplotlib_nomenclatura_03_0.jpg">

Por último, dentro de cada conjunto de ejes encontramos lo que vamos a llamar gráfica (referido con el nombre de **plot** en la literatura en inglés). Estas serían las cuatro gráficas generadas

## **plot**

El método **`plot()`**: crea una figura de forma **automatica y un conjunto de ejes**, y crea una grafica correspondiente con los datos que se le pasen como argumentos.

* `plot()` no visualiza en pantalla el gráfico, por ello se debe emplear la función `show()`

```python
plt.plot([2, 6, 3, 8])
plt.show()
```

## **subplots**

Nos ayuda a crear una figura y uno o más conjuntos de ejes (por defecto se crea uno).

```python
#creamos la funcion
x = np.arange(1,50,0.1)
y = np.sin(x)
#utilización del método subplots()
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

## **figure y axes**

Estos métodos, hacen de forma explicita y por separado, la figura y el conjunto de ejes

```python
fig = plt.figure()
ax = plt.axes()
ax.plot([2, 6, 7, 3, 8])
plt.show()
```

## **Interfaces de programación**

matplotlib, ofrece diferentes formas de conseguir el mismo objetivo (visualización)


**Interfaz estilo MATLAB (pyplot o pylab)**

El modulo **pylab** casi no se recomienda, pero es la interfaz mas similar a MATLAB.
Por otro lada el modulo pyplot, ofrece funciones y comandos similares a MATLAB para generar graficos.

Se recomienda por lo tanto utilizar la interfaz **pyplot** con el metodo **plot**, cuando uno comienza a utilizar matplotlib.

Pero tiene una desventaja, y lo que pasa es que con esta interfaz no se puede hacer referencia a una grafica (fig, axes) ya creada. Es decir si se crea una grafica 'A' y despues se crea la grafica 'B', y en un punto deseamos modificar la grafica 'A', esto ya no se lograr dado que la interfaz pyplot siempre tiene referencia al eje actual.

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.plot([2, 6, 7, 3, 8])
plt.title("Ventas")
# las marcas en los ejes (lo que en matplotlib se denominan ticks)
plt.xticks(range(0,5),["Ene", "Feb", "Mar", "Abr", "May"])
plt.subplot(1, 2, 2)
plt.hist(np.random.randint(1, 10, 100), bins=9)
plt.title("Almacenes")
plt.show()
```

**Interfaz orientada a Objetos(POO)**

Esta es la interfaz recomendada para los programadores experimentados, ya que nos permite tener siempre una referencia (variables) a objetos ya creados.

Aqui si se ocupa la interfaz pyplot, pero con el metodo** subplots** o **subplot2Grid**, los cuales nos devuelve una fiugura y una lista del conjunto de axes.

```python
#En este caso, las variables fig y ax nos servirán para personalizar todas las gráficas a crear en la figura
fig, ax = plt.subplots(1,2)
ax[0].plot([2, 6, 7, 3, 8])
ax[0].set_title("Ventas")
ax[0].set_xticks(range(0,5)) #Valores
ax[0].set_xticklabels(["Ene", "Feb", "Mar", "Abr", "May"]) #Etiquetas.
ax[1].hist(np.random.randint(1, 10, 100), bins=9)
ax[1].set_title("Almacenes")
plt.show()

#La mayor parte de los métodos que nos permiten fijar un valor a un atributo comienza por "set_". 
#Estos métodos suelen tener un método equivalente que comienza habitualmente por "get_" 
#que sirve para leer el valor de dicho atributo

```
