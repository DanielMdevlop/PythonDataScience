# **ESTILOS DE LÍNEA EN LA GRAFICA**

La funcion plot, nos permite configurar la grafica con cierto detalle vía parametros.

```
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## **Estilo y Ancho de Línea**


```
y = np.random.randn(50).cumsum()
w = np.random.randn(50).cumsum()
```

```
plt.plot(y)
plt.show()
```

Se aplicara formato a la grafica (NO A LA FIGURA O A LOS EJES)

**`linestyle`**: Permite especificar, el estilo de linea (por defecto es continua).

<img src="https://matplotlib.org/_images/sphx_glr_line_styles_reference_001.png"></img>

Mas información: 

https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html


```
plt.plot(y, linestyle="--")
plt.show()
```

```
plt.plot(y, linestyle="--")
plt.plot(w, linestyle="dashdot")
plt.show()
```

**`linewidth`**: Nos permite controlar el ancho de las lineas generadas. Acepta un numero real.

```
plt.plot(y, linewidth = 5)  # un ancho 5
plt.show()
```

```
plt.plot(y, linewidth = 5)  # un ancho 5
plt.plot(w, linestyle="dashdot", linewidth=2.5)
plt.show()
```

## **Marcadores y Colores**

Por defecto la funcion `plot()` dibuja los puntos (interseccion entre los valores de **x** e **y** ) con lineas rectas.

**`marker`**: nos permite mostrar un marcador encima de cada uno de los puntos con un simbolo (marcador)

```
plt.plot(y, marker = "o")
plt.show()
```

Para saber más sobre los marcadores consulte: [Pagina oficial](https://matplotlib.org/api/markers_api.html#module-matplotlib.markers)

Algunos ejemplos: 

<img src="https://www.interactivechaos.com/sites/default/files/inline-images/tutorial_matplotlib_plot_args_07.JPG" width="600" height="400">

Otros parametros son:

*  **`markeredgewidth`**: Ajusta el ancho del borde del marcador.
*  **`markeredgecolor`**: Permite controlar el color del borde del marcador. 
*  **`markerfacecolor`**: Ajusta el color de fondo del marcador. 
*  **`markerfacecoloralt`**: permite seleccionar un color alternativo para el color de fondo del marcador.
*  **`markersize`**: Permite controlar el tamaño del marcador

```
#Ejemplo del parametro markersize y markerfacecolor

plt.plot(y, marker = "o", markersize = 12, markerfacecolor = "red")
plt.show()
```

***color***: este parametro controla el color de la grafica y acepta diferentes formatos, consult: [Enlace](https://https://www.interactivechaos.com/manual/tutorial-de-matplotlib/marcadores-y-colores)
y aclare sus dudas en la seccion color.

Sin embargo, en este caso recomendamos los colores RGB y RGBA o los colores para diseño web.

```
#En este caso aplicamos un color de : https://colorhunt.co/
# donde se utiliza el formato hexadecimal.
plt.plot(y, color = "#28df99", linewidth = 3)
plt.show()
```

```
#EXISTE La otra forma es ocupar una letra, la cual representa un color ejemplo
#b = blue
#r = red, etc.
#tambien colores web como tomato
plt.plot(y, color = "b", linewidth = 3)
plt.show()
```
