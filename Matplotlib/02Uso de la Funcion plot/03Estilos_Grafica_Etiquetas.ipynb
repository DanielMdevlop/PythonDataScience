# **Estilos de gr√°fica y etiquetas.**

```
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

**`drawstyle`**: define el estilo de grafica

* default
* steps: genera una grafica escalonada.

```
y = np.random.randn(50).cumsum()
w = np.random.randn(50).cumsum()
```

```
plt.plot(y, drawstyle = "default", label = "default")
plt.legend()
plt.show()

plt.plot(y, drawstyle = "steps", label = "steps")
plt.legend()
plt.show()
```

**`label`**: asigna una etiqueta a la grafica, que sera mostrada en leyenda (si la activamos con la funcion` legend()`).

```
plt.plot(np.random.randn(50).cumsum(), label = "Gr√°fica 1", linestyle = "--" )
plt.plot(np.random.randn(50).cumsum(), label = "Gr√°fica 2", linewidth = 5)
plt.plot(np.random.randn(50).cumsum(), label = "Gr√°fica 3", marker = "o")
plt.legend() #Para visualizar la etiqueta, previamente definida en plot()
plt.show()

#Observamos las leyendas, donde con colores, sabemos de gr√°fica se trata cada una
```
