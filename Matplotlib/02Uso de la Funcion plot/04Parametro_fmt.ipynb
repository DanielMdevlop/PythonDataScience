# **EL Parametro [fmt] de la funcion plot**

```
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```
y = np.random.randn(50).cumsum()
```

Hasta el momento, se han visto los parametros necesarios para personalizar la grafica, sin embargo el parametro fmt es una alternativa interesante:

* fmt: es un parametro de tipo cadena texto.
* se puede incluir un color: b, r (blue, red)
* se puede especificar un marcador: o
* Asi como el estilo de linea con los caracteres: --


Para mas información: [Referencia](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.plot.html)


```
#Se fija el color verde, indicando el caracter g
plt.plot(y, "g")
plt.show()
```

```
#Ahora especificamos el marcador a usar con la letra o y estilo de linea discontinua 
#con los carteres --
plt.plot(y, "o--")
plt.show()
```

En ejemplo,indicamos que se cree una línea formada por guiones y puntos (caracteres "-."), amarilla (letra "y") y marcadores tipo "estrella" (carácter "*"). Adicionalmente utilizamos el parámetro markersize para fijar el tamaño del marcador

```
fig, ax = plt.subplots()
g1 = ax.plot(y, '-.y*', markersize = 15)
plt.show()
```

El parametro **`fmt `**tiene el siguiente formaro:

`fmt = '[marker][line][color]'`

otra combinación podria ser:

`[color][marker][line] `

Sin embargo, ningun parametro dentro de la cadena de texto es forzozo, asi que puede estar esta vacia.

## **El parametro [fmt], ¿En dos graficas dentro de la misma figura?**

```
y1 = np.random.randn(50).cumsum()
y2 = np.random.randn(50).cumsum()

#En este caso, estamos pasando solo un vector "y", pero por cada grafica diferente
#usamos el parametro fmt, como separador
#de esta forma podemos utilizar solo plot() y hacer muchas graficas en la misma figura

plt.plot(y1,"", y2)  #La grafica de y1, tiene el parametro fmt vacio y la grafica y2 se omite el parametro
plt.show()
```

```
y1 = np.random.randn(50).cumsum()
y2 = np.random.randn(50).cumsum()

#En el ejemplo mostrado a continuación se utilizan los parámetros [fmt] de las dos gráficas:

plt.plot(y1, "-g", y2, "-.r")
plt.show()
```
