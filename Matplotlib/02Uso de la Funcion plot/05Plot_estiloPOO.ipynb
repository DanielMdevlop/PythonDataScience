# **Funcion plot al estilo POO**

Para la generación de la figura y de uno (o más) conjuntos de ejes, utilizaremos la función subplots(), la cual crea objetos, con referencia a la figura y el conjunto de ejes ( si son mas de 2, se crea una lista de ejes)

`fig, ax = plt.subplots()`


Una vez tenemos la referencia al conjunto de ejes ax (supongamos que es solo uno), podemos crear una gráfica ejecutando el método plot sobre ax

```
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```
#El método plot devuelve una lista de objetos Line2D que representan la gráfica, objetos sobre los que es posible
#trabajar fijando sus propiedades.Por este motivo, almacenamos
#el resultado del método en una variable a la que llamamos g

fig, ax = plt.subplots()  #obtenemos una referencia
g = ax.plot(np.random.randn(50).cumsum())
print(len(g))
```

Por lo tanto, podemos aplicar formato al objeto "g", ejecutando metodos *equivalentes* a algunos de los argumentos de la función plot.

```
#Indexamos el objeto, asi hacemos referencia al conjunto de ejes, en este caso solo es 1 y recoremos que se comienza en 0
fig, ax = plt.subplots()  #obtenemos una referencia
g = ax.plot(np.random.randn(50).cumsum())
g[0].set_color("RoyalBlue") #Establecemos un color
g[0].set_linestyle("--")
g[0].set_linewidth(3)
plt.show()
```

```
#Ahora haremos el mismo ejemplo, pero gestionando los marcadores
fig, ax = plt.subplots()  #obtenemos una referencia
g = ax.plot(np.random.randn(50).cumsum())
g[0].set_color("Tomato") #Establecemos un color
g[0].set_marker("o") #Establecemos un marcador
g[0].set_markersize(8)
plt.show()
```

## **Creando dos ejes con subplots**

Recordemos que el metodo plot() devuelve una lista de objetos line2D que representan la grafica, Por lo tanto, gracias a esta lista de objetos podemos establecer estilo a la grafica; para saber mas sobre que atributos se tiene acceso visiste: [Line2D](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.lines.Line2D.html)

Por este motivo es siempre mejor guardar la lista del metodo plot() en una variable, para que se tenga un mejor control.

```
#Subplots: crea una matriz de conjuntos de ejes formada por una fila y dos columnas (es decir, dos conjuntos de ejes).
fig, ax = plt.subplots(1, 2)  #obtenemos dos referencias
y = np.random.randn(50).cumsum()

g = ax[0].plot(y)  # plot devuelve una lista de objetos Line2D que representan la gráfica
g[0].set_drawstyle("default")  #Opcion por defecto

print("Tipo de dato de g es: {}".format(type(g[0])))
f = ax[1].plot(y)
f[0].set_drawstyle("steps")

plt.show()
```

```
#Probando el atributo label


fig, ax = plt.subplots()  

g1 = ax.plot(np.random.randn(50).cumsum())  # plot devuelve una lista de objetos Line2D que representan la gráfica
g1[0].set_label("Grafica 1")

g2 = ax.plot(np.random.randn(50).cumsum())
g2[0].set_label("Grafica 2")

plt.legend()
plt.show()
```
