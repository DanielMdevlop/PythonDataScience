# **Creacion y perzonalización de figuras.**
La figura es el contenedor dentro del cual vamos a crear una o más gráficas

```
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```
y = np.random.randn(100).cumsum()
```

Los tres objetos (recordar que los objetos son instancias de una clase) que componen la figura: 

* **FigureCanvas**: Clase contenedora de la clase Figure
* **Figure**: Clase contenedora de la clase Axes.
* A**xes**: Es el area rectangular que contiene los elementos graficos (líneas, círculos, texto, etc.) Sera el elemento donde representaremos el gráfico como tal.

## **Creación de figuras**

```
#Crea una figura y nos devuelve la referencia a la misma
fig = plt.figure()
```

```
#Crea una figura con un conjunto de ejes (defualt 1), es decir nos da la referencia a la figura
fig, ax = plt.subplots()
```

En ambos casos, el resultado contenido en la variable fig es una referencia a un objeto de tipo matplotlib.figure.Figure, "contenedor" que engloba al resto de elementos involucrados en el dibujo de una gráfica

## **Personalización de figuras.**

La funcion **`figure()`**, acepta algunos argumentos, que definen propiedades visuales de la **figura** (NO DE LA GRAFICA).

```
#Cracion de una figura y despues se añade una grafica, dentro de ella.
fig = plt.figure()
plt.plot(y)
plt.show()
```

### **Tamaño y color de fondo**

**figsize**: Establece el ancho y el alto, en pulgadas, siendo el primer valor el ancho y el segundo el alto, dichos valores pueden ser una tupla o lista.

el valor por defecto es:` [6.4, 4.8]`

```
fig = plt.figure(figsize= [9, 6])
plt.plot(y)
plt.show()
```

**facecolor**: controla el color de fonde de la figura.

```
fig = plt.figure(facecolor = "LightSalmon")
plt.plot(y)
plt.show()
```

Comprobamos en la anterior imagen que una cosa es el fondo de la figura y otra el fondo de la gráfica.

### **Borde**

**edgecolor**: Establece color al borde de la figura (por defecto es blanco)

**linewidth**: Establece el ancho de dicho borde (por defecto es 0)

```
fig = plt.figure(edgecolor= "#2E8B57", linewidth= 10)
plt.plot(y)
plt.show()
```

### **Titulo**

**suptitle**: Permite establecer un título para la figura -no para las gráficas que estén contenidas en ella.

```
y1 = np.random.randn(100).cumsum()
y2 = np.random.randn(100).cumsum()
```

```
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(10, 4)
plt.suptitle("Ingresos y gastos 2020", fontsize=16)
ax[0].plot(y1)
ax[1].plot(y2)
plt.show()
```

Esta funcionalidad tambien esta disponible en el metodo homónimo de una figura.

```
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(10, 4)
fig.suptitle("Ingresos y gastos 2020", fontsize=16)
ax[0].plot(y1)
ax[1].plot(y2)
plt.show()
```

## **Personalización con estilo OO**

La mayor parte de los parametros estan también disponibles con métodos del objeto Figure.

* figsize → set_size_inches
* facecolor → set_facecolor
* edgecolor → set_edgecolor
* linewidth → No tiene equivalente

```
fig = plt.figure(linewidth= 10) #el ancho del borde ha tenido que ser impuesto vía parámetro, al no tener un equivalente en estilo OO
fig.set_size_inches(8, 3)
fig.set_facecolor([0.2, 0.4, 0.3, 0.2])
fig.set_edgecolor([0.2, 0.4, 0.3, 0.3])
plt.plot(y)
plt.show()
```
