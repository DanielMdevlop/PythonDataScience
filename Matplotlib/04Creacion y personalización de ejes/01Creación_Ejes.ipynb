# **Creación de ejes (AXES)**

```
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```
y = np.random.randn(100).cumsum()
```

## **Función subplots**

Crea una figura y uno (o varios) conjuntos de ejes, devolviendo una referencia a la figura y a los ejes, por defecto se crea un unico conjunto de ejes.

```
fig, ax = plt.subplots()
```

```
#ax es una referencia a un objeto de tipo AxesSubplot
type(ax)
```

Crear una matriz de conjuntos de ejes con dos fils y 3 columnas (6 conjuntos de ejes)

```
fig, ax = plt.subplots(3, 2)
fig.set_size_inches(8, 6)  # se ha aumentado el tamaño, para mejor visualización, 8 x 6 pulgadas c/U
```

```
type(ax)
```

```
ax.shape
```

```
ax #array de numpy, pero que contiene las referencias de cada AXES
```

Si queremos acceder a una conjunto en particular, indexamos como un array de numpy

```
fig, ax = plt.subplots(3, 2)
fig.set_size_inches(8, 6)
ax[1, 0].plot(y)
ax[1, 0].grid(True)  #hablitamos la rejilla
plt.show()
```

Los parametros **`sharex`** y **`sharey`**, controlan la compartición de las propiedades de los ejes. Por defecto el valor es False. Por ejemplo si queremos compartir el mismo eje horizontal en todos los conjuntos seria como sigue:

```
fig, ax = plt.subplots(2, 2, sharex=True)
fig.set_size_inches(8, 4)
ax[1, 0].plot(y)
plt.show()

#OBSERVE COMO Todos los conjuntos de ejes comparten las misma escala horizontal y se comparten, es decir los de arriba no estan rotulados
```

```
#PARTIENDO del ejemplo anterior hacemos algo analogo con el eje y
fig, ax = plt.subplots(2, 2, sharey=True)
fig.set_size_inches(8, 4)
ax[1, 0].plot(y)
plt.show()
```

```
fig, ax = plt.subplots(2, 2, sharex=True,sharey=True)
fig.set_size_inches(8, 4)
ax[1, 0].plot(y)
plt.show()

#Ahora se comparte ambos ejes, en todo el conjuntos de AXES.
```

**sharex y sharey**

* row: indicamos que cada fila de la matriz compartira el mismo eje x.
* col: de manera analoga pero con las columnas

## **Función axes**

Esta función añade un conjunto de ejes a la figura actual, y fija este nuevo conjunto de ejes como el actual.

```
fig = plt.figure()
ax = plt.axes() #creamos un conjunto de ejes y lo añadismos ala figura creada
ax.plot(y)
plt.show()
```

```
type(ax)
```

```
y1 = np.random.randn(100).cumsum()
y2 = np.random.randn(100).cumsum()
```

```
fig = plt.figure()

ax1 = plt.axes()
ax1.plot(y1)
plt.show()
```

Si en el bloque anterior de codigo, hubiermos agreado un segundo conjunto de ejes, especificando su posición y tamaño:

Para fijar la posición y tamaño de un eje podemos pasar una tupla a la función como primer argumento. Esta tupla contendrá la siguiente información:

**`[Posición izquierda, posición derecha, ancho, alto]`**


```
fig = plt.figure()

ax1 = plt.axes()
ax1.plot(y1)

ax2 = plt.axes([0.0, 0.0, 0.5, 0.5])
ax2.plot(y2, color = "tomato")
plt.show()
```

La segunda gráfica la hemos creado a partir del punto (0, 0) con ancho y alto de 0.5 (estos valores van referidos al tamaño de la figura. Es decir, un ancho de 1 equivale al ancho de la figura).

## **Funcion add_subplot**

El método asociado a una figura ***añade un conjunto de ejes a la misma*** pero suponiéndolo en una matriz de columnas y filas (de conjuntos de ejes) y creándolo en la posición indicada de dicha matriz.

```
y = np.random.randn(100).cumsum()
```

```
fig = plt.figure()
fig.add_subplot()
plt.plot(y)
plt.show()
```

Pero si se invoca añadiendo como argumentos el **número de filas de la matriz, el número de columnas de la matriz y la posición que ocupa el conjunto de ejes** a crear

```
y1 = np.random.randn(100).cumsum()
y2 = np.random.randn(100).cumsum()
y3 = np.random.randn(100).cumsum()
```

```
fig = plt.figure()
fig.add_subplot(2, 2, 1)
plt.plot(y1)

fig = plt.figure()
fig.add_subplot(2, 2, 2)
plt.plot(y2, color = "tomato")

fig = plt.figure()
fig.add_subplot(2, 2, 3)
plt.plot(y3, color = "yellow")

plt.show()
```
