Elementos que dependen del conjunto de ejes (AXES)

* Title: título a mostrar en la gráfica
* x axis label: nombre del eje x
* y axis label: nombre del eje y
* Major ticks: marcas principales en los ejes
* Minor ticks: marcas secundarios en los ejes
* Major tick labels: etiquetas a mostrar en las marcas principales de los ejes
* Minor tick labels: etiquetas a mostrar en las marcas secundarios de los ejes
* Spines: bordes de la gráfica
* Grid: rejilla a mostrar sobre la gráfica

```
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline
```

```
y = np.random.randn(100).cumsum()
```

## **Titulo**

con la función title(), podemos añadir un titulo como primer argumento, pero tambien acepta otros parametros: 

* **alpha**: transparencia del texto
* **backgroundcolor**: color de fondo
* **color**: color del texto
* **fontfamily**: familia de la fuente a usar
* **fontname**: nombre de la fuente a usar
* **fontsize**: tamaño de la fuente
* **fontstretch**: ancho de la fuente
* **fontstyle**: estilo de la fuente ('normal', 'italic', 'oblique')
* **fontvariant**: variante de la fuente ('normal', 'small-caps')
* **fontweight**: "peso" de la fuente ( 'ultralight', 'light', 'normal', 'regular', 'bold', etc.)
* **horizontalalignment**: alineación horizontal del texto ('center', 'right', 'left')
* **linespacing**: espacio entre líneas
* **position**: posición x e y del título
* **rotation**: ángulo de rotación del texto (en grados)
* **verticalalignment**: alineación vertical del texto ('center', 'top', 'bottom', 'baseline', 'center_baseline')
* **x**: posición x del texto
* **y**: posición y del texto


```
#Estilo MATLAB
plt.plot(y)
plt.title("Ventas 2020")
plt.show()
```

```
#estilo POO
fig, ax = plt.subplots()
ax.plot(y)
ax.set_title("Ventas 2020")
plt.show()
```

A continuación se especifica una fuente de tamaño 20 y alineamos el texto a la derecha (extremo derecho del texto se muestra en el centro del área disponible)

```
fig, ax = plt.subplots()
ax.plot(y)
ax.set_title("Ventas 2020", fontsize = 20, horizontalalignment = "right")
plt.show()
```

Ahora se configurar la posición del texto (posición x e y en la figura), su tamaño, color y peso.

```
fig, ax = plt.subplots()

plt.plot(y)
plt.title("Ventas 2020", position = (0.75, 0.85),
             fontsize = 18,
             color = "SteelBlue",
             fontweight = "bold")
plt.show()
```

```
#codigo en el estilo orientado a objetos

fig, ax = plt.subplots()

ax.plot(y)
title = ax.set_title("Ventas 2020")  #Retorna una instancia de title
#por eso tiene sentido llamarla titile y la llamada de los metodos, hacen mencion a la ocnfiguracion del titulo
title.set_position(xy = [0.75, 0.85])
title.set_fontsize(18)
title.set_color("SteelBlue")
title.set_fontweight("bold")
plt.show()
```

## **Etiquetas de ejes**

Es posible asignar etiquetas a los ejes x e y con las funciones xlabel()  y ylabel()

```
fig, ax = plt.subplots()
plt.plot(y)
plt.xlabel("Año en curso")
plt.ylabel("Millones de $")
plt.show()
```

Las funciones mencionadas, tambien se le pueden especificar atributos.

```
fig, ax = plt.subplots()
plt.plot(y)
plt.xlabel("Año en curso",
           fontsize = 20,
           color = "w",
           fontstretch = 0,
           backgroundcolor = "SteelBlue")

plt.ylabel("Millones de $",
           fontsize = 20,
           color = "w",
           fontstretch = 0,
           backgroundcolor = "SteelBlue")

plt.show()
```

Los metodos de los conjuntos de ejes tienen sus equivalente al estilo OOP.
ambos devolviendo una referencia al objeto texto creado.

```
fig, ax = plt.subplots()
ax.plot(y)
xlabel = ax.set_xlabel("Año en curso")
ylabel = ax.set_ylabel("Millones en $")
plt.show()
```

## **Marcas de ejes**

```
x = range(100, 200)
y = np.random.randn(100).cumsum()
```

```
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

matplotlib, trata de ajustar las escalas lo mas razonable posible. Ejemplo en la imagen anterior, el eje "x" cubre todo el rango de nuestros valores del vector x (definido previamente) en una escala de 20 en 20. Sin embargo esto siempre sudecera, ya sea con el eje horizontal o vertical.

Las funciones xticks y yticks: Nos permiten obtener y/o establecer las marcas (escalas) en los ejes.

```
fig, ax = plt.subplots()
ax.plot(x, y)
plt.xticks(range(120,180,10))
plt.show()

#EN este ejemplo:mostramos las marcas del eje x solo entre los valores 120 y 180(no inclusive), pero de 10 en 10
```

 SI quisiéramos que las marcas se mostrasen incluyendo valores de x no incluidos en nuestro rango de datos, por ejemplo entre 0 y 250 de 20 en 20

```
fig, ax = plt.subplots()
ax.plot(x, y)
plt.xticks(range(0,250,20))
plt.show()
```

Para el eje y el proceso es exactamente el mismo. Mostremos marcas, por ejemplo solo entre los valores 120 y 180 de 10 en 10 en el eje x, y entre -2 y 4 de 1 en 1 en el eje y

```
fig, ax = plt.subplots()
ax.plot(x, y)
plt.xticks(range(120, 180, 10))
plt.yticks(range(-2,4))
plt.show()
```

Las funciones xticks y yticks nos permiten fijar marcas en los ejes, sin embargo tambien podemos sustituir dichas marcas por etiquetas.

Ejemplo: las etiquetas del eje x: son {120, 130, .., etc} -> se pueden fijar por
{2010, 2011, ..., etc}

```
xticks_labels = [2010, 2011, 2012, 2013, 2014, 2015]
fig, ax = plt.subplots()
ax.plot(x, y)
plt.xticks(range(120, 180, 10), xticks_labels)
plt.yticks(range(-2,4))
plt.show()
```

Los funciones mostradas tiene su metodo equivalente POO:
* plt.xticks → ax.set_xticks()
* plt.yticks → ax.set_yticks()

```
xticks_labels = [2010, 2011, 2012, 2013, 2014, 2015]
fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_xticks(range(120,180,10))
ax.set_xticklabels(xticks_labels) #método para establecer solo la etiqueta
ax.set_yticks(range(-2, 4))
plt.show()
```

## **Limites de los ejes**

```
y = np.random.randn(100).cumsum()
```

```
np.min(y), np.max(y)
```

```
fig, ax = plt.subplots()
ax.plot(y)
plt.show()

#Comprobamos como matplotlib dimenciona ambos ejes de forma automatica,
#y con ello cubre el rango de valores
```

Si ahora queremos mostrar una parte especifica de la grafica, por ejemplo entre los valores 20-60 en el eje x., esto lo logramos con la funcion xlim()

```
fig, ax = plt.subplots()
ax.plot(y)
plt.xlim(20, 60) #esto se puede ver como un ZOOM a la grafica.
plt.show()
```

De forma equivalente tenemos la funcion **ylim** para eje vertical

```
fig, ax = plt.subplots()
ax.plot(y)
plt.ylim(-2, 4)
plt.show()
```

En el estilo OO

```
fig, ax = plt.subplots()
ax.plot(y)
ax.set_xlim(20, 60)
ax.set_ylim(-2, 4)
plt.show()
```

```
fig, ax = plt.subplots()
ax.plot(y)
ax.set_xlim(20, 60)
ax.set_ylim(-2, 4)
x = ax.get_xlim()
plt.show()
print("los limites de x es: ",x)
```

```
fig, ax = plt.subplots(1,2)
fig.set_size_inches(11, 4)
ax[0].plot(y)
ax[1].plot(y)
ax[1].set_ylim(-2, 4)
plt.show()
```

## **Relación de aspecto**

Podemos personalizar de los ejes es su relación de aspecto (la relación entre una unidad en el eje x y una unidad en el eje y)

```
y = np.random.randn(100).cumsum()
```

```
fig, ax = plt.subplots()
ax.plot(y)
plt.show()
```

```
# equal tiene la relacion de aspecto que sera de 1:1

fig, ax = plt.subplots()
ax.plot(y)
ax.set_aspect("equal")
plt.show()
```

```
fig, ax = plt.subplots()
ax.plot(y)
ax.set_aspect(2)
plt.show()
```
