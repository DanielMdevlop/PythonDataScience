```
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## **Grid**

Es la famosa rejilla, que sirve como referencia para facilitar la interpretación de los datos.

```
y = np.random.randn(100).cumsum()
```

```
fig, ax = plt.subplots()
ax.plot(y)
ax.grid()
plt.show()
```

Parametros que permiten configurar el grid

* **alpha**: grado de transparencia
* **color**: color de las líneas del grid
* **linestyle**: estilo de las líneas ('-', '--', '-.', ':', etc.)
* **linewidth**: ancho de las líneas

Ejemplo:


```
fig, ax = plt.subplots()
ax.plot(y)
ax.grid(alpha = 0.5,
        color = "black",
        linestyle = "-.",
        linewidth = 1.5)
plt.show()
```
