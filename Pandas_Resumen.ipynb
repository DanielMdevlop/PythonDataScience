## **¬øQu√© es un DataFrame?**

Una estructura tabular o tabla, se compone por filas y columnas, donde cada una de las columnas puede admitir diferentes tipos de datos, sin embargo, nunca habra dos tipos de datos en la misma columna.

Cada fila y cada columna tiene dos tipos de identificadores. 

*  Los indices posicionales ( son los secuenciales empezando por cero)
*  Las etiquetas (normalmente son las columnas las que tienen etiqueta)

## **Limpieza de datos**

En el caso que no se quiera eliminar los campos vacios o nulos, estos se pueden

## **Transformaci√≥n de Datos**

```
#Agregando columnas para transformar los datos, es decir agregar mas informacion para el dataset

#Primero se debe definir el proceso para agregar esa columna
import random
def calcularaGanancias(retweets):
    return retweets * random.randint(3, 5)
```

```
df["ganancias"] = df["retweets"].apply(calcularaGanancias)

df.head()
```

```
def popularidad(fila):
    return fila["followees"] / fila["followers"]
```

```
df["popularidad"] = df.apply(popularidad, axis=1)

df.head()
```

## **Agrupar los datos**

Se pueden agrugar por datos por medio de una categoria.

Agrupar significa unir filas que tengan un mismo valor dentro de una columna. Pero ¬øQue sucede con los valores de otras columnas que se se agrupan?
Estos valores deben sufrir una transformaci√≥n. Que puede ser por ejemplo, quedarnos con el promedio de todos los valores que se agrupan, o coger el valor m√°ximo o el minimo o sumar todos los valores.

Algo similar que se hace en estadistica con la tablas de frecuencia agrupadas.

```
#Agrupar el dataframe por la columna pais
df.head()
```

Se agrupa con el metodo `groupby`
Ahora la columna country se ha puesto en negritas, esto quiere decir que dicha columna ha pasado a ser nuevo identificador de mi DataFrame

```
 
df.groupby("country").mean()

#Observamos que solo tenemos una fila por cada pais, y a todas la columnas se les ha aplicado una funcion promedio
```

```
#Aplicar diferentes funciones a cada columna de la agrupaci√≥n:

grouped = df.groupby("country").agg({"followers" : 'sum', 
                           "mentions" : 'mean', 
                           "retweets" : 'max'
                          })
```

```
grouped[grouped["followers"] > 5000]
```
